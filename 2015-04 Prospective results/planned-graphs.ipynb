```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
x = np.arange(11)

f, ax1 = plt.subplots()
freq = ax1.plot(x, 5 + (9 - 5) * (1 - np.exp(-x/3)))[0]
ax1.set_xlabel("Sentence depth in tree")
ax1.set_ylabel("Log-freqency", color=freq.get_color())
for t in ax1.get_yticklabels():
    t.set_color(freq.get_color())
ax1.grid()

ax2 = ax1.twinx()
age = ax2.plot(x, 11 + (8 - 11) * (1 - np.exp(-x/5)),
               color=sb.color_palette()[1])[0]
ax2.set_ylabel("Age of Acquisition", color=age.get_color())
for t in ax2.get_yticklabels():
    t.set_color(age.get_color())
ax2.grid()

f.savefig('freq-age-tree-evolution.png', dpi=300)
```

```python
f, ax = plt.subplots()
items = ax.plot(x, [1, .8, .8, .7, .6, .6, .6, .4, .4, .4, .3], color=sb.color_palette()[2])[0]
ax.set_xlabel("Sentence depth in tree")
ax.set_ylabel("Proportion of initial items maintained", color=items.get_color())
for t in ax.get_yticklabels():
    t.set_color(items.get_color())
f.savefig('items-tree-evolution.png', dpi=300)
```

```python
import pandas as pd
from numpy import random

genders = ['male', 'female']
CSPs = ["Professions Intermédiaires",
        "Employés",
        "Ouvriers"]

n_authors = 100
gGenders = random.choice(genders, size=n_authors)
gCSPs = random.choice(CSPs, size=n_authors)
gBiases = random.normal(size=n_authors)

authors = pd.DataFrame({'gender': gGenders,
                        'bias': gBiases,
                        'CSP': gCSPs})

g = sb.factorplot("CSP", "bias", "gender", authors, kind="box",
                   palette="PRGn", aspect=1.25, x_order=CSPs)
g.despine(offset=10, trim=True)
g.set_axis_labels("CSP", "1st LDA component bias")
g.savefig('author-bias.png', dpi=300)
```
